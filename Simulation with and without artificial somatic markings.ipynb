### Calcular la distancia

Se calcula distancias de la matriz de distancia para obtener las distancia en kilómetros y en minutos que se recorren en una ruta

```python
def distancia (a, datos):
    suma=0
    for i in range(len(a)-1):
        suma= suma + datos.loc[a[i]].loc[a[i+1]]
        
    return (suma)
```

### Puntos turísticos 

Se cuenta la cantidad de lugares turísticos por lo que pasa la ruta seleccionada. 

```python
def puntos (a):
    suma=0
    i=0
    for i in range(len(a)):
        b=a[i]
        if b > 29:
            suma= suma + 1
        
    return (suma)
```
